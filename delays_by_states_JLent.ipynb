```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from citipy import citipy
import warnings
import holoviews as hv
import hvplot.pandas
#import API key
#from api_keys import geoapify_key
warnings.filterwarnings('ignore')
```

```python
#flight data to dataframe
flight_data_df = pd.read_csv("flights_sample_3m.csv")
flight_data_df.head()
```

```python
#separate flight date, airline, flight number, origin city, dest city, cancelled
simplified_flight_df = flight_data_df[["FL_DATE","AIRLINE","ORIGIN_CITY","DEST_CITY","DEP_DELAY","CANCELLED"]]
simplified_flight_df
```

```python
#Create new column called Origin State from Origin_City based on ","
simplified_flight_df["ORIGIN_STATE"] = simplified_flight_df["ORIGIN_CITY"].str.rsplit(',').str[-1]

#Display the first 5 rows of the DataFrame
simplified_flight_df.head()
```

```python
#filter the dataframe to include only rows with positive dep_delays
filtered_df = simplified_flight_df[(simplified_flight_df["DEP_DELAY"] > 0)]
```

```python
#Create new column in the DataFrame called Dest_State from Dest_City based on ","
filtered_df["DEST_STATE"] = filtered_df["DEST_CITY"].str.rsplit(',').str[-1]

#Display the first 5 rows
filtered_df.head()
```

```python
#convert FL_DATE to datetime and extract year
filtered_df["FL_DATE"] = pd.to_datetime(filtered_df["FL_DATE"])
filtered_df["YEAR"] = filtered_df["FL_DATE"].dt.year

#number of flights per state per year
flights_per_state_yr = filtered_df.groupby(["YEAR", "ORIGIN_STATE"]).size()

#convert groupby object to DataFrame
flights_per_state_yr_df = flights_per_state_yr.reset_index().rename(columns={0: "NUM_FLIGHTS"})

#show number of flights per state per year
flights_per_state_yr_df
```

```python
#calculate mean for flight delays per year
mean_delays = filtered_df.groupby("YEAR")["DEP_DELAY"].mean()

#plot mean delay times
mean_delays.plot(kind="bar", figsize=(12,6))
plt.ylabel("Mean Delay Time (mins)")
plt.title("Mean Flight Delay Time Per Year")
plt.show
```

```python
#calculate mean for flight delays per state
mean_delays_state = filtered_df.groupby("ORIGIN_STATE")["DEP_DELAY"].mean()

#convert the series to a dataframe
mean_delays_state_df = mean_delays_state.to_frame()

#rename column
mean_delays_state_df.columns = ["Mean Delay"]

#sort the dataframe in descending order
mean_delays_state_df = mean_delays_state_df.sort_values(by="Mean Delay",ascending=False)

#plot mean delay times
mean_delays_state_df.plot(kind="bar", figsize=(12,6))
plt.ylabel("Mean Delay Time (mins)")
plt.title("Mean Flight Delay Time Per State")
plt.show
```

```python
mean_delays_state_df
```

```python
#avg departure delay per state per year
avg_dep_delay_state_yr = filtered_df.groupby(["YEAR", "ORIGIN_STATE"])["DEP_DELAY"].mean()

#convert series to DataFrame
avg_dep_delay_state_yr_df = avg_dep_delay_state_yr.reset_index()

#get unique years and states
years = avg_dep_delay_state_yr_df["YEAR"].unique()
states = avg_dep_delay_state_yr_df["ORIGIN_STATE"].unique()

#create figure and axis
fig, axs = plt.subplots(len(years),1,figsize=(15,5*len(years)))

#for each year create bar plot
for i, year in enumerate(years):
    #filter data for the year
    data_for_year = avg_dep_delay_state_yr_df[avg_dep_delay_state_yr_df["YEAR"] == year]

    #sort the data in descending order
    data_for_year = data_for_year.sort_values(by="DEP_DELAY", ascending=False)

    #create bar plot
    axs[i].bar(data_for_year["ORIGIN_STATE"], data_for_year["DEP_DELAY"])
    axs[i].set_title(f"Average Departure Delay Per State In {year}")
    axs[i].set_xlabel("State")
    axs[i].set_ylabel("Average Departure Delay")

#display plots
plt.tight_layout()
plt.show()
```

```python
# Calculate median delay time per state per year
median_delays = filtered_df.groupby(["YEAR", "ORIGIN_STATE"])["DEP_DELAY"].median()

# Convert Series to DataFrame
median_dep_delay_state_yr_df = median_delays.reset_index()

# Get unique years and states
years = median_dep_delay_state_yr_df["YEAR"].unique()
states = median_dep_delay_state_yr_df["ORIGIN_STATE"].unique()

# Create figure and axes
fig, axs = plt.subplots(len(years), 1, figsize=(15, 5 * len(years)))

# For each year, create a bar plot
for i, year in enumerate(years):
    # Filter data for the year
    data_for_year = median_dep_delay_state_yr_df[median_dep_delay_state_yr_df["YEAR"] == year]

    

    # Create bar plot
    axs[i].bar(data_for_year["ORIGIN_STATE"], data_for_year["DEP_DELAY"])
    axs[i].set_title(f"Median Departure Delay Per State In {year}")
    axs[i].set_xlabel("State")
    axs[i].set_ylabel("Median Departure Delay")

# Display plots
plt.tight_layout()
plt.show()
```

```python
#calculate mode delay time per state
mode_delays = filtered_df.groupby(["YEAR", "ORIGIN_STATE"])["DEP_DELAY"].agg(lambda x: x.mode()[0])

#show mode delay times
#mode_delays

fig, axs = plt.subplots(len(years), 1, figsize=(10, 5 * len(years)))

for i, year in enumerate(years):
    data_for_year = filtered_df[filtered_df["YEAR"] == year]
    data_for_year.set_index("ORIGIN_STATE", inplace=True)
    data_for_year["DEP_DELAY"].sort_values().plot(ax=axs[i], kind='line')
    
    axs[i].set_title(f"Mode Departure Delay Per State In {year}")
    axs[i].set_xlabel("State")
    axs[i].set_ylabel("Mode Departure Delay")

plt.tight_layout()
plt.show()
```
